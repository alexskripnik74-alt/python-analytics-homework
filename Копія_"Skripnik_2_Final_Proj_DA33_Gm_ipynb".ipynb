```python
import pandas as pd

# Використовуємо префікс 'r' перед рядком (raw string), щоб уникнути помилок зі шляхами Windows
path = r'd:\DA 33\DA Fin Project\stack-overflow-developer-survey-2025'

# Завантажуємо основні дані
df = pd.read_csv(f'{path}\survey_results_public.csv')

# Завантажуємо схему (опис структури)
schema_df = pd.read_csv(f'{path}\survey_results_schema.csv')

# Перевіряємо, чи все завантажилось (виведемо перші 5 рядків основних даних)
df.head()
```

```python
# Завдання 1. Підрахунок загальної кількості респондентів 1. Перевірка завантаження та підрахунок
total_respondents = df.shape[0]

# 2. Виведення результату
print(f"Загальна кількість респондентів: {total_respondents}")

# Відображення просто як число (як того вимагає формат здачі)
total_respondents
```

```python
# Завдання 2 а. Отримуємо список назв запитань зі схеми
survey_questions = set(schema_df['qname'])

# Отримуємо список усіх колонок, які є в основному датасеті
dataset_columns = set(df.columns)

# Використовуємо інтерсекцію множин, щоб залишити лише ті назви, які є і в схемі, і в даних
target_columns = list(survey_questions.intersection(dataset_columns))

print(f"Кількість питань для перевірки: {len(target_columns)}")
```

```python
# Завдання 2 б. Створюємо підмножину даних лише з колонками-запитаннями
df_questions_only = df[target_columns]

#  Використовуємо метод dropna(), щоб видалити рядки, де є хоча б один пропуск (NaN)
# how='any' означає: видалити рядок, якщо хоча б одне питання пропущено
respondents_with_all_answers = df_questions_only.dropna(how='any')

#  Підраховуємо кількість - Отримуємо фінальне число
total_complete_responses = len(respondents_with_all_answers)

# Виведення результату у заданому форматі
print(f"кількість респондентів, які відповіли на всі запитання з опитування. : {total_complete_responses}")
```

```python
# Завдання 3 1. Розрахунок мір центральної тенденції
mean_val = df['WorkExp'].mean()
median_val = df['WorkExp'].median()
mode_val = df['WorkExp'].mode()[0]  # .mode() повертає Series, беремо перший елемент

# 2. Створення зручного звіту через DataFrame для гарного відображення
stats_report = pd.DataFrame({
    'Показник': ['Середнє значення', 'Медіана', 'Мода'],
    'Значення (роки)': [mean_val, median_val, mode_val]
})

# Виведення результату
print("Аналіз досвіду роботи респондентів (WorkExp):")
stats_report.set_index('Показник')
```

```python
# Завдання 4 1. Визначаємо назву колонки (RemoteWork) та фільтруємо значення 'Remote'
remote_workers_count = df[df['RemoteWork'] == 'Remote'].shape[0]

# 2. Виведення фінального результату
print(f"Кількість респондентів, які працюють віддалено: {remote_workers_count}")

# Відображення числа для звіту
remote_workers_count
```

```python
# Завдання 5 1. Визначаємо колонку з мовами
language_col = 'LanguageHaveWorkedWith'

# 2. Підраховуємо загальну кількість респондентів, які відповіли на це питання
# (виключаємо тих, хто не вказав жодної мови)
total_answered = df[language_col].dropna().shape[0]

# 3. Фільтруємо респондентів, у яких в списку мов є 'Python'
# na=False ігнорує порожні клітинки, щоб не було помилки
python_users_count = df[df[language_col].str.contains('Python', na=False)].shape[0]

# 4. Розраховуємо відсоток
python_percentage = (python_users_count / total_answered) * 100

# 5. Виведення результату у відсотковому форматі
print(f"Відсоток респондентів, які програмують на Python: {python_percentage:.1f}%")

# Відображення чистого значення для результату
round(python_percentage, 1)
```

```python
# Завдання 6 1.Визначаємо назву колонки для навчання
learning_col = 'LearnCode'

# 2. Шукаємо варіант "Online Courses" у колонці.
# Використовуємо str.contains, оскільки респонденти часто обирають кілька варіантів навчання.
# na=False ігнорує порожні значення.
online_learners_count = df[df[learning_col].str.contains('Online Courses', na=False)].shape[0]

# 3. Виведення фінального результату
print(f"Кількість респондентів, які навчалися через онлайн-курси: {online_learners_count}")

# Відображення числа для звіту
online_learners_count
```

```python
import pandas as pd

# Завдання  7 1. Фільтруємо респондентів, які програмують на Python
python_devs = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

# 2. Групуємо за країною та обчислюємо середнє і медіану
country_comp = python_devs.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

# 3. Перейменовуємо колонки
country_comp = country_comp.rename(columns={
    'mean': 'Середня компенсація',
    'median': 'Медіанна компенсація'
})

# 4. Видаляємо країни, де немає даних про компенсацію (NaN)
country_comp = country_comp.dropna()

# 5. Форматуємо обидві колонки: один знак після коми та роздільник тисяч (кома)
# Використовуємо '{:,.1f}', де ',' — роздільник тисяч, '.1f' — один знак після коми
country_comp['Середня компенсація'] = country_comp['Середня компенсація'].map('{:,.1f}'.format)
country_comp['Медіанна компенсація'] = country_comp['Медіанна компенсація'].map('{:,.1f}'.format)

# 6. Сортуємо (зверніть увагу: після форматування дані стають рядками,
# тому сортуємо до того, як застосувати формат, або за числовим показником)
# Але в даному випадку виведемо топ-15 країн з найвищими медіанами
country_comp.head(15)
```

```python
#   Завдання 8 1. Створюємо копію топ-5 найбільш оплачуваних, видаляючи пропуски
top_5_paid = df.dropna(subset=['ConvertedCompYearly']).sort_values(by='ConvertedCompYearly', ascending=False).head(5).copy()

# 2. Вибираємо розширений набір колонок для аналізу
result = top_5_paid[['ConvertedCompYearly', 'EdLevel', 'Country', 'YearsCode']].copy()

# 3. Форматуємо компенсацію (перетворюємо на текст із роздільниками)
result['ConvertedCompYearly'] = result['ConvertedCompYearly'].astype(float).map('{:,.1f}'.format)

# 4. Виведення фінальної таблиці
print("Аналіз топ-5 найбільш оплачуваних респондентів:")
result
```

```python
#  Завдання 9 1. Створюємо допоміжну колонку, яка вказує, чи знає респондент Python (True/False)
df['Uses_Python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

# 2. Групуємо за віком та розраховуємо середнє значення для логічної колонки.
# В Python середнє від True/False (1/0) — це і є частка (відсоток, якщо помножити на 100).
age_analysis = df.groupby('Age')['Uses_Python'].mean() * 100

# 3. Перетворюємо результат у DataFrame для гарного вигляду
age_analysis_df = age_analysis.reset_index()
age_analysis_df.columns = ['Вікова категорія', 'Відсоток Python-розробників']

# 4. Округлюємо відсотки до одного знака після коми та додаємо символ %
age_analysis_df['Відсоток Python-розробників'] = age_analysis_df['Відсоток Python-розробників'].map('{:.1f}%'.format)

# Виведення результату
print("Популярність Python за віковими категоріями:")
age_analysis_df
```

```python
# 1. Розрахунок 75-го перцентиля для компенсації
# Використовуємо dropna, щоб розрахунок базувався лише на реальних числах
q75 = df['ConvertedCompYearly'].quantile(0.75)

# 2. Фільтрація за двома критеріями:
# - Компенсація вища за 75-й перцентиль
# - Формат роботи — 'Remote'
high_paid_remote = df[
    (df['ConvertedCompYearly'] > q75) &
    (df['RemoteWork'] == 'Remote')
].copy()

# 3. Підрахунок популярності індустрій
# Примітка: в опитуванні 2024/2025 за індустрію зазвичай відповідає колонка 'Industry'
industry_counts = high_paid_remote['Industry'].value_counts()

# 4. Формування фінальної таблиці
industry_report = industry_counts.reset_index()
industry_report.columns = ['Індустрія', 'Кількість респондентів']

print(f"Поріг 75-го перцентиля: {q75:,.1f}")
print("Топ індустрій серед високооплачуваних віддалених працівників:")
industry_report
```
